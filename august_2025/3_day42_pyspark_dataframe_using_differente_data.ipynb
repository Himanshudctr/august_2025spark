# create datafrmae list of values with schema

## list based schema

```python
list_data = [(1,'sreeni', 30),(2,'ram',35),(3,'raju',40)]
schema = ['id', 'name', 'age']       # list based schema
df2=spark.createDataFrame(data = list_data, schema = schema)
df2.show()
```

```python
df2.printSchema()
```

## String based Shame inculding dtype

```python
list_data = [(1,'sreeni', 30.3),(2,'ram',35.2),(3,'raju',40.4)]
schema = "id int, name string, age float"  # sting based schema
df3=spark.createDataFrame(data = list_data, schema = schema)
df3.show()
```

```python
df3.printSchema()
```

### display() vs df.show()

display() → Works only in Databricks, shows data in table format, supports download, 
and you can click the ➕ (plus sign) to see Profiling and Visualization options.

df.show() → Works everywhere, shows data in text format only, no download or visuals.

✅ Use display(df) in Databricks for data analysis, profiling, and export.

✅ Use df.show() in scripts or local environments for compatibility.
