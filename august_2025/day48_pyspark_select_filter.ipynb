### day_48_pyspark_select_filter

```python
data = [
    {"customer_id": 1, "name": "Alice", "age": 30, "city": "New York"},
    {"customer_id": 2, "name": "Bob", "age": 25, "city": "San Francisco"},
    {"customer_id": 3, "name": "Charlie", "age": 35, "city": "Chicago"}
]

df = spark.createDataFrame(data)

df.show()
```

```python
df.createOrReplaceTempView("df_v")
```

```python
spark.sql("select customer_id,name from df_v").display()
```

```python
from pyspark.sql.functions import col

display(
    df.select(
        col("city").alias("city name"),
        col("name")
    )
)
```

```python
from pyspark.sql.functions import col
df.select(col("customer_id"),col("name")).display()
```

here multiple syntace use

```python
df.select(col("customer_id"), df['name'], df.city.alias('city name'), 'age').display()
```

```python
from pyspark.sql.functions import col, upper, lower

df.select(
    col('age').alias('person_age'),
    col('customer_id').alias('customer_id'),
    col('name')
).display()
```

```python
spark.sql('select customer_id, age, age+10, upper(name) from df_v ').display()
```

```python
df.select(col('customer_id'),
          col('age'),
          col('age')+10,
          upper(col('name')).alias('lastname')).display()
```

```python
df.filter(df.age > 30).display()

df.select('*').filter(col('age') > 30).display()
```

```python
df1 = spark.read.csv("/Volumes/workspace/default/august_2025/Contact_info.csv", header=True, inferSchema=True)

# Display the data (optional)
display(df1)

# Now print only the column names
print(df1.columns)
```

```python
from pyspark.sql.functions import col, upper, trim, concat_ws

df1.select(
    col("Identifier"),
    col("Surname").alias("firstname"),
    col("`given name`").alias("lastname"),
    concat_ws("*",
              upper(col("Surname")),
              trim(upper(col("given name")))
             ).alias("full_name")
).display()
```

```python
df.select(df.colRegex("`^.*name*`")).show(2)
```

```python
from pyspark.sql.functions import col, lower, upper, concat_ws

df.select(
    col("name"),
    concat_ws("-", lower(col("name")), upper(col("name"))).alias("formatted_name")  # example only
).display()
```
