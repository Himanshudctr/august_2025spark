It only permits approved data sources (e.g., snowflake, mysql, sqlserver, redshift, delta, parquet, etc.).

You also can’t attach custom JARs on Serverless.

## Snowflake connection

```python
df = (
    spark.read.format("snowflake")
    .option("sfURL", "pgkktza-qw01651.snowflakecomputing.com")
    .option("sfUser", "AUGUSTAUTOMATION")
    .option("sfPassword", "Dharmavaram@2025")
    .option("sfDatabase", "test_db")
    .option("sfSchema", "test_schema")
    .option("dbtable", "SAMPLE_TABLE")
    .option("sfWarehouse", "<YOUR_WAREHOUSE>")
    .load()
)
display(df)
df.printSchema()
```
